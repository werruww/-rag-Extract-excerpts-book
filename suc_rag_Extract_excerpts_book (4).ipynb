```python
from llama_index.readers.file import PyMuPDFReader
import pymupdf4llm
from langchain.text_splitter import MarkdownTextSplitter
import pathlib
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import FAISS

# ========= 1. تحويل PDF إلى نص Markdown =========
file_path = "1.pdf"
loader = PyMuPDFReader()
documents = loader.load(file_path=file_path)

# تحويل إلى Markdown
md_text = pymupdf4llm.to_markdown(file_path)

# حفظ النص في ملف
output_file = "output.md"
pathlib.Path(output_file).write_bytes(md_text.encode("utf-8"))

print(f"تم حفظ النص في {output_file}")

# ========= 2. تقسيم النص إلى أجزاء =========
splitter = MarkdownTextSplitter(chunk_size=1000, chunk_overlap=100)  # يمكن تعديل الحجم
split_docs = splitter.split_text(md_text)

print(f"تم تقسيم النص إلى {len(split_docs)} جزء")

# ========= 3. إنشاء قاعدة بيانات FAISS للبحث =========
vectorstore = FAISS.from_texts(split_docs, OpenAIEmbeddings())

# حفظ قاعدة البيانات لاستخدامها لاحقًا
vectorstore.save_local("faiss_index")

print("تم إنشاء قاعدة بيانات البحث بنجاح!")

# ========= 4. سؤال الكتاب =========
def ask_book(question):
    # تحميل قاعدة البيانات
    vectorstore = FAISS.load_local("faiss_index", OpenAIEmbeddings())

    # البحث عن الجواب
    docs = vectorstore.similarity_search(question)

    # طباعة الإجابات المحتملة
    print("\n🔹 الإجابات المحتملة:\n")
    for doc in docs:
        print(doc.page_content)
        print("-" * 50)

# مثال: طرح سؤال
question = "Author of the book"
ask_book(question)
```

```python
from llama_index.readers.file import PyMuPDFReader
import fitz  # مكتبة PyMuPDF لاستخراج النصوص بدقة
from langchain.text_splitter import MarkdownTextSplitter
import pathlib
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

# ========= 1. تحسين استخراج النص من PDF =========
file_path = "1.pdf"

def extract_text_from_pdf(pdf_path):
    """استخراج النص من PDF بدقة باستخدام PyMuPDF."""
    doc = fitz.open(pdf_path)
    text = "\n".join([page.get_text("text") for page in doc])  # استخراج النص من كل صفحة
    return text

# استخراج النص الخام
raw_text = extract_text_from_pdf(file_path)

# تحسين جودة النص عبر إزالة التكرار والمسافات الزائدة
cleaned_text = "\n".join([line.strip() for line in raw_text.split("\n") if line.strip()])

# حفظ النص في ملف
output_file = "output.txt"
pathlib.Path(output_file).write_bytes(cleaned_text.encode("utf-8"))

print(f"✅ تم حفظ النص النقي في {output_file}")

# ========= 2. تقسيم النص إلى أجزاء =========
splitter = MarkdownTextSplitter(chunk_size=800, chunk_overlap=100)  # زيادة الدقة
split_docs = splitter.split_text(cleaned_text)

print(f"✅ تم تقسيم النص إلى {len(split_docs)} جزء")

# ========= 3. تحسين دقة Embeddings =========
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")  # دقة أعلى

# إنشاء قاعدة بيانات FAISS
vectorstore = FAISS.from_texts(split_docs, embedding_model)

# حفظ قاعدة البيانات لاستخدامها لاحقًا
vectorstore.save_local("faiss_index")

print("✅ تم إنشاء قاعدة بيانات البحث بنجاح!")

# ========= 4. تحسين دقة البحث =========
def ask_book(question, top_k=5):
    # تحميل قاعدة البيانات مع السماح بفك pickle
    vectorstore = FAISS.load_local("faiss_index", embedding_model, allow_dangerous_deserialization=True)

    # البحث عن أفضل الإجابات مع ضبط `top_k`
    docs = vectorstore.similarity_search(question, k=top_k)

    # طباعة الإجابات مع درجات التشابه
    print("\n🔹 الإجابات المحتملة:\n")
    for i, doc in enumerate(docs, 1):
        print(f"🔹 نتيجة {i}:\n{doc.page_content}\n")
        print("-" * 50)

# مثال: طرح سؤال
question = "Author of the book"
ask_book(question, top_k=5)  # إرجاع 5 إجابات بدلاً من 1 فقط
```

 كما هو متوقع: استخراج النص، تقسيمه، إنشاء قاعدة بيانات للبحث، وإجراء البحث. ومع ذلك، يبدو أن السؤال "Author of the book" لم يُرجع إجابة مباشرة باسم المؤلف من النص المستخرج من ملف PDF "1.pdf".

تحليل المشكلة المحتملة:

محتوى ملف PDF: السبب الأكثر احتمالاً هو أن ملف PDF "1.pdf" نفسه قد لا يحتوي على إجابة مباشرة وواضحة لسؤال "Author of the book" بالطريقة التي تتوقعها. قد يكون الكتاب لا يذكر المؤلف بشكل صريح في الأجزاء التي تم استرجاعها كنتائج بحث.

طبيعة السؤال: سؤال "Author of the book" هو سؤال مباشر جدًا. إذا كان النص المستخرج لا يحتوي على جملة مثل "This book is written by [اسم المؤلف]" أو معلومات مماثلة، فقد لا يتمكن نموذج البحث الدلالي من العثور على إجابة مباشرة.

نتائج البحث: النتائج التي تم إرجاعها تدور حول مواضيع فيزيائية مثل "Einstein" و "relativity theory" و "quantum mechanics". هذا يشير إلى أن الكتاب على الأرجح يتناول مواضيع فيزيائية وقد يذكر شخصيات مثل أينشتاين، لكنه لا يقدم بالضرورة اسم المؤلف بشكل مباشر في هذه الأجزاء.

خطوات مقترحة للتحقق وتحسين النتائج:

فحص ملف PDF "1.pdf":

تأكد من وجود اسم المؤلف في الملف: افتح ملف PDF "1.pdf" وابحث يدويًا عن اسم المؤلف. هل هو مذكور في صفحة العنوان، المقدمة، أو أي جزء آخر من الكتاب؟

تحقق من سياق ذكر المؤلف: إذا كان اسم المؤلف مذكورًا، فهل هو مذكور في سياق يمكن لنظام البحث الدلالي فهمه كإجابة لسؤال "من هو مؤلف الكتاب؟".

تحسين السؤال:

جرب أسئلة أكثر تحديدًا: بدلاً من "Author of the book"، جرب أسئلة مثل:

"Who is the author of this book?" (من هو مؤلف هذا الكتاب؟)

"What is the name of the author?" (ما هو اسم المؤلف؟)

"By whom was this book written?" (بواسطة من كتب هذا الكتاب؟)

ابحث عن عبارات مفتاحية: ابحث عن عبارات مثل "written by" (كتب بواسطة)، "authored by" (من تأليف)، "by" (بواسطة) متبوعة باسم.

زيادة عدد النتائج (top_k):

جرب زيادة top_k: قمت بالفعل بضبط top_k=5. يمكنك محاولة زيادته إلى قيمة أعلى مثل top_k=10 أو top_k=15 لرؤية ما إذا كانت النتائج الإضافية تحتوي على معلومات حول المؤلف. ولكن كن حذرًا، زيادة top_k بشكل كبير قد يؤدي إلى نتائج أقل دقة.

تحسين معالجة النص (إذا لزم الأمر):

التأكد من جودة استخراج النص: PyMuPDF جيد جدًا، ولكن في حالات نادرة، قد تكون هناك مشاكل في استخراج النص من بعض ملفات PDF المعقدة. إذا كنت تشك في ذلك، يمكنك محاولة استخدام مكتبة أخرى لاستخراج النص للمقارنة (على الرغم من أن PyMuPDF هو الأفضل للدقة).

معالجة إضافية للنص: في حالات نادرة، قد تحتاج إلى إجراء معالجة إضافية للنص المستخرج، مثل إزالة التنسيقات المعقدة أو الجداول إذا كانت تتداخل مع البحث الدلالي (ولكن في هذه الحالة، يبدو النص المستخرج نظيفًا بشكل معقول).

باختصار:

يبدو أن الكود يعمل بشكل جيد. المشكلة على الأرجح تكمن في أن ملف PDF "1.pdf" قد لا يحتوي على إجابة مباشرة لسؤالك "Author of the book" في الأجزاء التي تم استرجاعها كنتائج بحث. الخطوة الأولى هي فحص ملف PDF نفسه والتحقق من وجود اسم المؤلف وكيفية ذكره. ثم جرب تحسين السؤال أو زيادة عدد النتائج.

```python
from llama_index.readers.file import PyMuPDFReader
import fitz  # مكتبة PyMuPDF لاستخراج النصوص بدقة
from langchain.text_splitter import MarkdownTextSplitter
import pathlib
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

# ========= 1. تحسين استخراج النص من PDF =========
file_path = "1.pdf"

def extract_text_from_pdf(pdf_path):
    """استخراج النص من PDF بدقة باستخدام PyMuPDF."""
    doc = fitz.open(pdf_path)
    text = "\n".join([page.get_text("text") for page in doc])  # استخراج النص من كل صفحة
    return text

# استخراج النص الخام
raw_text = extract_text_from_pdf(file_path)

# تحسين جودة النص عبر إزالة التكرار والمسافات الزائدة
cleaned_text = "\n".join([line.strip() for line in raw_text.split("\n") if line.strip()])

# حفظ النص في ملف
output_file = "output.txt"
pathlib.Path(output_file).write_bytes(cleaned_text.encode("utf-8"))

print(f"✅ تم حفظ النص النقي في {output_file}")

# ========= 2. تقسيم النص إلى أجزاء =========
splitter = MarkdownTextSplitter(chunk_size=800, chunk_overlap=100)  # زيادة الدقة
split_docs = splitter.split_text(cleaned_text)

print(f"✅ تم تقسيم النص إلى {len(split_docs)} جزء")

# ========= 3. تحسين دقة Embeddings =========
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")  # دقة أعلى

# إنشاء قاعدة بيانات FAISS
vectorstore = FAISS.from_texts(split_docs, embedding_model)

# حفظ قاعدة البيانات لاستخدامها لاحقًا
vectorstore.save_local("faiss_index")

print("✅ تم إنشاء قاعدة بيانات البحث بنجاح!")

# ========= 4. تحسين دقة البحث =========
def ask_book(question, top_k=5):
    # تحميل قاعدة البيانات مع السماح بفك pickle
    vectorstore = FAISS.load_local("faiss_index", embedding_model, allow_dangerous_deserialization=True)

    # البحث عن أفضل الإجابات مع ضبط `top_k`
    docs = vectorstore.similarity_search(question, k=top_k)

    # طباعة الإجابات مع درجات التشابه
    print("\n🔹 الإجابات المحتملة:\n")
    for i, doc in enumerate(docs, 1):
        print(f"🔹 نتيجة {i}:\n{doc.page_content}\n")
        print("-" * 50)

# مثال: طرح سؤال
question = "PHYSICS OF LIFE by "
ask_book(question, top_k=1)  # إرجاع 5 إجابات بدلاً من 1 فقط
```
